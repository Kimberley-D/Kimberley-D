<h2>Q. 9) Find all instances when 'Snow' was recorded.</h2>

There are three ways in which we can find the answer to the question above;

```python
# value_counts() which gives us a count of 390
#df.head(2)
df['Weather Condition'].value_counts()
```

```python
# Filtering gives us the same answer of 390 rows/instances
df[df['Weather Condition'] == 'Snow']
```

```python
# str.contains to count all instances where snow was recorded regardless of any other weather conditions that were recorded which gives us a total count of 583
df[df['Weather Condition'].str.contains('Snow')]

# We can use df[df['Weather Condition'].str.contains('Snow')].head(50) to explore the first 50 occurances of snow
```
